#Tring to find the best model for datasets

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean') #to impute the missing values
```

```python
imputer=imputer.fit(df) # fitting the dataset with imputed values
```

```python
df=imputer.transform(df)
```

```python
df
```

```python
df=pd.DataFrame(df)
df
```

```python
y=df.iloc[:,-1] # using iloc to check the value of particular row/coloumn
```

```python
y.head() # top 5 values
```

```python
y.shape
```

```python
x=df.iloc[:,1:-1]
```

```python
x.head()
```

```python
x.shape
```

```python
# as dataset is now ready now we will train and test the model
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# now we use Liner Regression so will run the model with .fit function-
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
lm.coef_
```

```python
# now we will predict the value of x_test
pred=lm.predict(x_test)
```

```python
print('predicted value and actual value is',pred,y_test)
```

```python
# now we will check the aaccurcy of model while finding MSE and MAE
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('Mean absolute error:',mean_absolute_error(y_test,pred))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
```
